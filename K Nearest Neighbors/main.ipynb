## Write a program to develop the KNN classifier for the k values as 3,5,7
## based on split up of training and testing dataset as 90-10, 70-30 and leaveone-out on,
### a) Glass dataset
### b) Fruit dataset
### using the different distance metrics like Euclidean and Manhattan distance.

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
glass_dataset = pd.read_csv('glass.csv')
glass_dataset
```

```python
fruits_dataset = pd.read_csv('fruits.csv')
fruits_dataset
```

```python
datasets = ['glass_dataset', 'fruits_dataset']
# The values are vector representations, which is required for the knn algorithm 
Xg, yg = glass_dataset.drop('Type', axis=1).values, glass_dataset['Type'].values
Xf, yf = fruits_dataset[['mass', 'width', 'height', 'color_score']].values, fruits_dataset['fruit_label'].values
k_values = [3,5,7]
distance_metrics = ['manhattan', 'euclidean']
test_splits = [0.1, 0.3]
```

```python
for dataset in datasets:
    print(f'Dataset: {dataset}')
    print('----------------------')
    for k in k_values:
        for p, distance_metric in enumerate(distance_metrics):
            for test_split in test_splits:
                print(f'Distance Metric: {distance_metric}')
                print(f'K-Value: {k}')
                print(f"Train-Test Split = {100 - test_split * 100}-{test_split * 100}")
                
                model = KNN(n_neighbors=k, p=p+1, weights="distance")
                
                if dataset == 'glass_dataset':
                    X_train, X_test, y_train, y_test = train_test_split(Xg, yg, test_size=test_split, random_state=50)
                else:
                    X_train, X_test, y_train, y_test = train_test_split(Xf, yf, test_size=test_split, random_state=50)
                    
                model.fit(X_train, y_train)
                accuracy = model.score(X_test, y_test)
                print(f'Accuracy: {accuracy: .2f}')
                print("------------")
            print("----------------------------")
    print("------------------------------------------------")
```
