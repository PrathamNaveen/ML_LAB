## Write a program to perform unsupervised K-means clustering techniques

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris # This is the iris dataset in numerical format
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = load_iris().data
```

```python
k = 3
model = KMeans(n_clusters=k)
model.fit(dataset)
labels = model.labels_
centroids = model.cluster_centers_
```

```python
centroids
```

```python
plt.scatter(dataset[:,0], dataset[:,1], c=labels) # All rows 0th Column and All rows 1st Column
plt.scatter(centroids[:,0], centroids[:,1], marker='x', color='red', s=200)
plt.title('K-Means Clustering')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
```
