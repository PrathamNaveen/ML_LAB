## Write a program to develop a decision tree classifier based on weather
## forecasting dataset.

```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier as DecisionTree
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('weather_forecast.csv')
dataset
```

```python
desired_rows= 1500
dataset = pd.concat([dataset] * (desired_rows//len(dataset)), ignore_index=True)
dataset.shape
```

```python
# Convert to numerical format
for col in dataset.columns:
    dataset[col] = dataset[col].map({x : (i+1) for i, x in enumerate(dataset[col].unique()) })
dataset.head()
```

```python
X = dataset.drop('Play', axis=1).values
y = dataset['Play'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)

model = DecisionTree(max_depth=3)
model.fit(X_train, y_train)
print(tree.export_text(model, max_depth=3, feature_names=list(dataset.columns[:-1]), class_names=['No', 'Yes']))
print("Accuracy: ", model.score(X_test, y_test))
```
